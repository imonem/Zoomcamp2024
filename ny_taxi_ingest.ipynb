## Using [pyarrow](https://arrow.apache.org/docs/python/parquet.html) to import parquet

*parquet is the standard for big data*

```python
import pandas as pd
# import pyarrow.parquet as pq
```

```python
# df = pd.read_parquet('yellow_tripdata_2021-01.parquet')
df = pd.read_csv('taxi+_zone_lookup.csv')
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
engine.connect()
```

```python
print(pd.DataFrame.to_sql(df, con=engine, name='taxi_zone_lookup', chunksize=100000, if_exists="append"))
```

```python
df.head()
```

```python
df.head(n=0).to_sql(name='taxi_zone_lookup', con=engine, if_exists='replace')
```

```python
query = """
SELECT 1 as number;
"""

pd.read_sql(query, con=engine)
```

```python
query = """
SELECT *
FROM pg_catalog.pg_tables
WHERE schemaname != 'pg_catalog' AND schemaname != 'information_schema';
"""

pd.read_sql(query, con=engine)
```

```python
query = """
SELECT *
FROM yellow_taxi_data
LIMIT 10;
"""

pd.read_sql(query, con=engine)
```
